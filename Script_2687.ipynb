```
#Mengimport Library
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```
#Mengupload data dan menampilakn data pada urutan 5 teratas 
data = pd.read_excel("online_retail.xlsx")
data.head()
```

```
#Menampilakan kolom data
data.columns
```

```
#Menampilkan kolom Country pada data dengan fungsi unique agar tidak muncul ouput yang sama/double 
data.Country.unique()
```

```
#Menbersihkan data
#Mengapus runag ekstra pada deskripsi
data['Description'] = data['Description'].str.strip()

#Menghapus baris tanpa nomor faktur(Invoice)
data.dropna(axis= 0, subset=['Invoice'], inplace= True)
data['Invoice']= data['Invoice'].astype('str')

#Menghapus semua transaksi yang dilakukan secara kredit
data= data[~data['Invoice'].str.contains('C')]
```

```
#Memisahkan data berdasarkan dengan negara teransaksi
#Traksaksi yang dilakukan di Prancis
basket_France = (data[data['Country'] =="France"]
		.groupby(['Invoice', 'Description'])['Quantity']
		.sum().unstack().reset_index().fillna(0)
		.set_index('Invoice'))
basket_France
```

```
#Mendifinisikan fungsi hot encoding supaya data lebih mudah untuk disesuaikan
def hot_encode(x):
    if(x<= 0):
        return 0
    if(x>= 1):
        return 1
```

```
#Encoding dataset
basket_encoded = basket_France.applymap(hot_encode)
basket_France = basket_encoded
basket_France
```

```
#Membuat model apiori
frq_items = apriori(basket_France, min_support = 0.05, use_colnames = True)

#Mengumpulkan aturan yang disimpulakan dalam dataframe
rules = association_rules(frq_items, metric ="lift", min_threshold = 1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
rules.head(2)
```
